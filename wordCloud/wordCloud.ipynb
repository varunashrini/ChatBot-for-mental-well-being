Extremely Positive

```python
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS

# Read text file
with open("0.txt", "r") as file:
    text = file.read()

# Generate word cloud
word_cloud = WordCloud(
    width=3000,
    height=2000,
    random_state=1,
    background_color="white",
    colormap="Pastel1",
    collocations=False,
    stopwords=STOPWORDS,
).generate(text)

# Display the generated Word Cloud
plt.imshow(word_cloud)
plt.axis("off")
plt.show()
  
```
